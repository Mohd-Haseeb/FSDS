### Working with mysql

```python

def create_connection(host_name, user_name, password, database):

    try:
        connection = mysql.connector.connect(host=host_name, user = user_name, password=password, use_pure=True, auth_plugin='mysql_native_password')
    except Error as err:
        print('Error : ',err)
    else:
        return connection
        
def execute_query(connection, query:str,commit:bool=False):
    cursor = connection.cursor()
    result = None
    try:
        cursor.execute(query)
        result = cursor.fetchall()
        if commit:
            connection.commit()
        print("Query was successfully executed")
    except Error as err:
        print('Error : ',err)
    else:
        return result
    

def close_connection(connection):
    # cursor.close()
    connection.close()



        
        
```

### If we consider `Distance` as a criteria for finding Lazy person. Here, we will find percentage of lazy distance a user walks when compared with Total Distance.

> NOTE : Here we are considering (SedentaryActiveDistance+LightActiveDistance) as a LAzy Distance

```python
id_distance = df_fitbit.groupby('Id').aggregate(
    {'LightActiveDistance':'sum','SedentaryActiveDistance':'sum','VeryActiveDistance':'sum','ModeratelyActiveDistance':'sum'}
    ).reset_index()
```

```python
id_distance.head(8)
```

```python
lazy_distance = (id_distance.iloc[:,1] + id_distance.iloc[:,2])
total_distance = id_distance.iloc[:,1] + id_distance.iloc[:,2] + id_distance.iloc[:,4] + id_distance.iloc[:,4] 

id_distance['lazy_distance_percentage'] = ((lazy_distance/total_distance))*100
```

```python
id_distance.sample(6)
```

```python
id_distance.sort_values(['lazy_distance_percentage'], ascending=False)[:5]
```

-> Users with Id `2026352035` and `2891001357` has highest percentage in terms of Lazy Distance Percentage
